```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('/content/Iris.csv')
df.head()
```

```python
df.describe()
```

```python
x_var = df.iloc[:,1:5]
x_var
```

```python
y_var = df.iloc[:,5]
y_var
```

```python
sc = StandardScaler()
trans = sc.fit_transform(x_var)
trans
```

```python
trans.T
```

```python
cov_matrix=np.cov(trans.T)
cov_matrix
```

```python
eigen_values,eigen_vectors=np.linalg.eig(cov_matrix)
eigen_values
```

```python
eigen_pairs=[(np.abs(eigen_values[i]),eigen_vectors[:,i]) for i in range(len(eigen_values))]
print('Eigen values in descending order')
for i in eigen_pairs:
  print(i[0])
```

```python
pca = PCA()
pca = pca.fit(trans)
```

```python
explained_var = pca.explained_variance_ratio_
explained_var
```

```python
plt.plot(np.cumsum(explained_var))
plt.xlabel('Number of principal components')
plt.ylabel('Cumulative explained variance')
plt.show()
```

```python
pca_2 = PCA(n_components= 4 )
pca_2 = pca_2.fit(trans)
pca_2d = pca_2.transform(trans)
pca_2d
```

```python
plt.scatter(pca_2d[:,0],pca_2d[:,1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
np.dot(pca_2.components_[0],pca_2.components_[1])
```

```python
cov_mat=np.cov(pca_2d,rowvar=False)
cov_mat
```
