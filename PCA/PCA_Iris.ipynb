```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Load dataset
data = sns.load_dataset("iris")
data.head()
X = data.drop(columns=["species"])
y = data["species"]

# Train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
data.head()
```

```python
# Without PCA
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model = LogisticRegression(max_iter=200)
model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test_scaled)
acc_without_pca = accuracy_score(y_test, y_pred)
```

```python
# With PCA(1 components)
pca = PCA(n_components=1)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

model.fit(X_train_pca, y_train)
y_pred_pca1 = model.predict(X_test_pca)
acc_with_pca1 = accuracy_score(y_test, y_pred_pca1)

print("Explained Variance:", pca.explained_variance_ratio_)
```

```python
# With PCA(2 components)
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

model.fit(X_train_pca, y_train)
y_pred_pca2 = model.predict(X_test_pca)
acc_with_pca2 = accuracy_score(y_test, y_pred_pca2)
```

```python
# With PCA(3 components)
pca = PCA(n_components=3)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

model.fit(X_train_pca, y_train)
y_pred_pca3 = model.predict(X_test_pca)
acc_with_pca3 = accuracy_score(y_test, y_pred_pca3)
```

```python

print("Explained Variance:", pca.explained_variance_ratio_)
print("Accuracy without PCA:", acc_without_pca)
print("Accuracy with PCA (1 component):", acc_with_pca1)
print("Accuracy with PCA (2 component):", acc_with_pca2)
print("Accuracy with PCA (3 component):", acc_with_pca3)
```
